## Matplotlibについて
***
matplotlibは、プログラミング言語Python及びその科学計算用ライブラリNumpyのためのグラフ描画ライブラリである。オブジェクト指向のAPIを提供しており、様々な種類のグラフを描画する能力を持つ。描画できるのは主に2次元のグラフを描画する(プロット)だが、3次元のグラフを描画する(プロット)機能も追加されてきている。

## MAaplotlibできること
***
- 折線グラフと散布図の描画
- グラフを並べて表示するsubplotsメソッド
- 円グラフ
- 棒グラフ
<br>
以上のデータの可視化をすることが可能。

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.DataFrame(np.random.rand(5, 5), index=['A', 'B', 'C', 'D', 'E'], columns=['A1', 'B2', 'C3', 'D4', 'E5'])
```

```python
df
```

```python
# 散布図
# plt.scatter(df['カラム名'], df['カラム名'])

plt.scatter(df['A1'], df['B2'])

# 一方の値(x)が増加する時に、他方の値(y)が減少する関係を「負の相関」と言う
```

```python
# ヒストグラム
# plt.hist(df['カラム名'])
# ヒストグラムでは「データの散らばり」を確認することができる
# binsという幅を指定することで広く狭くデータを確認することができる
# binsのデフォルトは10

plt.hist(df['C3'], bins=20)

# np.array([])とヒストグラムが表示される
# 下記では0.2という値が1.0個あるのが確認できる
```

```python
# 折線グラフ
# np.linspaceは指定した値をx等分することが可能
# (x, y)のようなグラフを作成する

x = np.linspace(1, 100, 100)
x
```

```python
y = 2*x
y
```

```python
plt. plot(x, y)
plt.grid()
```

```python
y_1 = 1/x
y_1
```

```python
# 反比例のグラフを作成することも可能

plt.plot(x, y_1)
plt.grid()
```

```python
# 二次関数
y_2 = 5*x*x
y_2
```

```python
plt.plot(x, y_2)
plt.grid()
```

```python
# subplot
# subplotで複数のグラフを並べてみる

np.linspace(-np.pi, np.pi, 201)
```

```python
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.title('sin & cos')
```

```python
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(121)
# カラー措定とラベル(凡例)
plt.plot(x, y_sin, color='yellow', label='sin')
# タイトル
plt.title('sin')

plt.subplot(122)
# カラー措定とラベル(凡例)
plt.plot(x, y_cos, color='red', label='cos')
plt.title('cos')

# 横のスペース指定
plt.subplots_adjust(wspace=0.5)

# 全体の大きさ
plt.figure(figsize=(9, 3))

# グラフ上部の位置変更
plt.subplots_adjust(top=0.8)
```

```python
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(121)
plt.plot(x, y_sin)

plt.subplot(222)
plt.plot(x, y_cos)

plt.subplot(224)
plt.plot(x, y_1)

plt.suptitle('sin & cos & Inverse proportional', fontsize=18)
plt.subplots_adjust(wspace=0.5)
plt.grid()
plt.show()
```

```python
# 円グラフ

# ラベル指定
labels = ['A', 'B', 'C', 'D']
# サイズ指定
sizes = [15, 30, 35, 10]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels)

plt.show()
```

```python
# 円グラフ切り離し

# 円グラフ

# ラベル指定
labels = ['A', 'B', 'C', 'D']
# サイズ指定
sizes = [15, 30, 35, 10]
# 切り離し
explode = (0, 0.1, 0.05, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, explode=explode)

plt.show()
```

```python
# 円グラフ時計回り

# 円グラフ

# ラベル指定
labels = ['A', 'B', 'C', 'D']
# サイズ指定
sizes = [15, 30, 35, 10]
# 切り離し
explode = (0, 0.1, 0.05, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, explode=explode, counterclock=False)

plt.show()
```

```python
# 円グラフ影をつける

# 円グラフ

# ラベル指定
labels = ['A', 'B', 'C', 'D']
# サイズ指定
sizes = [15, 30, 35, 10]
# 切り離し
explode = (0, 0.1, 0.05, 0)

fig1, ax1 = plt.subplots()
# 影をつける(shadow=True)
ax1.pie(sizes, labels=labels, explode=explode, counterclock=False, shadow=True)

plt.show()
```

```python
# 値を円グラフ内に表示

# 円グラフ

# ラベル指定
labels = ['A', 'B', 'C', 'D']
# サイズ指定
sizes = [15, 30, 35, 10]
# 切り離し
explode = (0, 0.1, 0.05, 0)

fig1, ax1 = plt.subplots()
# 値を設定(autopct)
ax1.pie(sizes, labels=labels, explode=explode, counterclock=False, shadow=True, autopct='%1.1f%%')

plt.show()
```

```python
# ドーナツ

# 円グラフ

# ラベル指定
labels = ['A', 'B', 'C', 'D']
# サイズ指定
sizes = [15, 30, 35, 10]
# 切り離し
explode = (0, 0.1, 0.05, 0)
# ドーナツサイズ
size = 0.5

fig1, ax1 = plt.subplots()
# ドーナツ状へ設定(wedgeprop)
ax1.pie(sizes, labels=labels, explode=explode, counterclock=False, shadow=True, autopct='%1.1f%%', wedgeprops=dict(width=size, edgecolor='w'))

plt.show()
```

```python
# 半径

# 円グラフ

# ラベル指定
labels = ['A', 'B', 'C', 'D']
# サイズ指定
sizes = [15, 30, 35, 10]
# 切り離し
explode = (0, 0.1, 0.05, 0)
# 半径(円型)サイズ
size = 0.5

fig1, ax1 = plt.subplots()
# 半径設定(radius)
ax1.pie(sizes, labels=labels, explode=explode, counterclock=False, shadow=True, autopct='%1.1f%%', wedgeprops=dict(width=size, edgecolor='w'), radius=2)

plt.show()
```

```python
# 入れ子の円グラフ

# 円グラフ

# ラベル指定
labels = ['A', 'B', 'C', 'D']
# サイズ指定
sizes = [15, 30, 35, 10]
# 切り離し
explode = (0, 0.1, 0.05, 0)
# 半径(円型)サイズ
size = 0.5

fig1, ax1 = plt.subplots()
# 半径設定(radius)
ax1.pie(sizes, labels=labels, explode=explode, counterclock=False, shadow=True, autopct='%1.1f%%', wedgeprops=dict(width=size, edgecolor='w'), radius=2)

plt.show()
```
